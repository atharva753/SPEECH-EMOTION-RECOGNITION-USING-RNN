```python
# Install required libraries (if not already installed)
!pip install numpy librosa

# Import necessary libraries
import os
import librosa
import numpy as np
import csv

# ----------------------
# STEP 1: Set Up Paths
# ----------------------
# Update this variable to point to the folder where your RAVDESS dataset is stored in Google Drive.
# For example, if you've mounted your Google Drive and uploaded the dataset to "MyDrive/RAVDESS", use:
dataset_root = '/content/drive/MyDrive/kaggle/RAVDESS'

# Create an output folder for extracted features and CSV file.
output_folder = '/content/drive/MyDrive/ser/extracted_features'
os.makedirs(output_folder, exist_ok=True)

# -----------------------------
# STEP 2: Define Emotion Mapping
# -----------------------------
emotion_labels = {
    '01': 'Neutral',
    '02': 'Calm',
    '03': 'Happy',
    '04': 'Sad',
    '05': 'Angry',
    '06': 'Fearful',
    '07': 'Disgust',
    '08': 'Surprised'
}

# --------------------------------------
# STEP 3: Create CSV Mapping & Extract Features
# --------------------------------------
csv_filename = os.path.join(output_folder, 'labels.csv')
with open(csv_filename, mode='w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(['feature_file', 'label'])  # CSV header

    # Loop over all files in the dataset folder
    for subdir, dirs, files in os.walk('/content/drive/MyDrive/kaggle/RAVDESS'):
        for file in files:
            if file.lower().endswith('.wav'):
                file_path = os.path.join(subdir, file)

                # Extract emotion code from the filename
                # Adjust index (here: parts[2]) if your dataset naming convention is different.
                parts = file.split('-')
                if len(parts) < 3:
                    continue  # Skip files with unexpected format
                emotion_code = parts[2]
                if emotion_code not in emotion_labels:
                    continue  # Skip if emotion code is not recognized
                label = emotion_labels[emotion_code]

                # Load audio file
                try:
                    audio, sr = librosa.load(file_path, sr=None)
                except Exception as e:
                    print(f'Error loading {file_path}: {e}')
                    continue

                # Extract MFCC features; adjust n_mfcc if needed.
                n_mfcc = 40
                mfcc = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=n_mfcc)

                # Save the MFCC features as a .npy file in the output folder
                feature_filename = os.path.splitext(file)[0] + '.npy'
                feature_filepath = os.path.join(output_folder, feature_filename)
                np.save(feature_filepath, mfcc)

                # Write the mapping (filename and label) to the CSV file
                csv_writer.writerow([feature_filename, label])
                print(f'Processed {file} -> {feature_filename} (Label: {label})')

print('Feature extraction completed!')
```
